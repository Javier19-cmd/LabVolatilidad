Eventos recientes del sistema

```python
import win32evtlog

# Conectar al registro de eventos del sistema
server = 'localhost'
logtype = 'System'  # Otras opciones: 'Application', 'Security'
hand = win32evtlog.OpenEventLog(server, logtype)

# Leer los eventos
flags = win32evtlog.EVENTLOG_FORWARDS_READ | win32evtlog.EVENTLOG_SEQUENTIAL_READ
total = win32evtlog.GetNumberOfEventLogRecords(hand)

events = win32evtlog.ReadEventLog(hand, flags, 0)
for event in events:
    print(f"Evento: {event.EventID}, Fecha: {event.TimeGenerated}, Fuente: {event.SourceName}")
```

Programas instalados

```python
import winreg

# Abrir la clave de programas instalados
key = winreg.OpenKey(winreg.HKEY_LOCAL_MACHINE, r"SOFTWARE\Microsoft\Windows\CurrentVersion\Uninstall")
i = 0
while True:
    try:
        subkey_name = winreg.EnumKey(key, i)
        subkey = winreg.OpenKey(key, subkey_name)
        display_name = winreg.QueryValueEx(subkey, "DisplayName")[0]
        print(display_name)
        i += 1
    except OSError:
        break
```

Aplicaciones instaladas

```python
import winreg

def list_installed_software():
    reg_path = r"SOFTWARE\Microsoft\Windows\CurrentVersion\Uninstall"
    key = winreg.OpenKey(winreg.HKEY_LOCAL_MACHINE, reg_path)
    
    i = 0
    while True:
        try:
            subkey_name = winreg.EnumKey(key, i)
            subkey = winreg.OpenKey(key, subkey_name)
            try:
                app_name = winreg.QueryValueEx(subkey, "DisplayName")[0]
                print(f"Aplicaci√≥n instalada: {app_name}")
            except FileNotFoundError:
                pass
            i += 1
        except OSError:
            break

list_installed_software()
```

Monitoreo de uso de CPU y memoria en tiempo real

```python
import psutil
import time

def monitor_system(duration):
    start_time = time.time()
    while True:
        # Verifica si ha pasado el tiempo especificado
        elapsed_time = time.time() - start_time
        if elapsed_time > duration:
            break
        
        # Uso de CPU
        cpu_usage = psutil.cpu_percent(interval=1)
        # Uso de memoria
        memory_info = psutil.virtual_memory()
        
        print(f"Uso de CPU: {cpu_usage}%")
        print(f"Memoria usada: {memory_info.used / (1024 ** 3):.2f} GB")
        print(f"Memoria disponible: {memory_info.available / (1024 ** 3):.2f} GB")
        print(f"Porcentaje de uso de memoria: {memory_info.percent}%")
        
        time.sleep(2)  # Esperar 2 segundos entre las actualizaciones

# Define el tiempo de monitoreo en segundos (10 segundos)
monitor_duration = 10
monitor_system(monitor_duration)
```

Extracci√≥n de configuraciones de red y conexiones activas

```python
import psutil

def network_info():
    # Obtener las interfaces de red y su configuraci√≥n
    addrs = psutil.net_if_addrs()
    for interface, addr_list in addrs.items():
        print(f"Interfaz: {interface}")
        for addr in addr_list:
            print(f"  Direcci√≥n: {addr.address}")
            print(f"  Tipo: {addr.family}")
    
    # Obtener conexiones activas
    connections = psutil.net_connections(kind='inet')
    print("\nConexiones activas:")
    for conn in connections:
        laddr = f"{conn.laddr.ip}:{conn.laddr.port}"
        raddr = f"{conn.raddr.ip}:{conn.raddr.port}" if conn.raddr else "N/A"
        print(f"  Local: {laddr} -> Remoto: {raddr} ({conn.status})")

network_info()
```

Monitoreo de uso del disco (Lectura/escritura) en tiempo real

```python
import psutil
import time

def monitor_disk_usage(duration):
    start_time = time.time()
    while True:
        elapsed_time = time.time() - start_time
        if elapsed_time > duration:
            break

        # Obtener las estad√≠sticas de lectura/escritura de disco
        disk_io = psutil.disk_io_counters()

        print(f"Bytes le√≠dos: {disk_io.read_bytes / (1024 ** 2):.2f} MB")
        print(f"Bytes escritos: {disk_io.write_bytes / (1024 ** 2):.2f} MB")
        print(f"Lecturas completadas: {disk_io.read_count}")
        print(f"Escrituras completadas: {disk_io.write_count}")
        
        time.sleep(2)  # Esperar 2 segundos entre actualizaciones

# Monitorear el uso del disco por 15 segundos
monitor_disk_usage(15)
```

Monitoreo del uso de red (Env√≠o/Recepci√≥n de datos)

```python
import psutil
import time

def monitor_network_usage(duration):
    start_time = time.time()
    while True:
        elapsed_time = time.time() - start_time
        if elapsed_time > duration:
            break
        
        # Obtener las estad√≠sticas de la red
        net_io = psutil.net_io_counters()

        print(f"Bytes enviados: {net_io.bytes_sent / (1024 ** 2):.2f} MB")
        print(f"Bytes recibidos: {net_io.bytes_recv / (1024 ** 2):.2f} MB")
        print(f"Paquetes enviados: {net_io.packets_sent}")
        print(f"Paquetes recibidos: {net_io.packets_recv}")
        
        time.sleep(2)  # Esperar 2 segundos entre actualizaciones

# Monitorear el uso de red por 15 segundos
monitor_network_usage(15)
```

Monitoreo de los archivos abiertos por un proceso aleatorio

```python
import psutil
import random

def list_open_files_random_process():
    # Obtener todos los procesos en ejecuci√≥n
    processes = list(psutil.process_iter())

    # Seleccionar un proceso al azar
    if processes:
        random_process = random.choice(processes)
        try:
            open_files = random_process.open_files()
            pid = random_process.pid
            name = random_process.name()

            print(f"Proceso seleccionado aleatoriamente: {name} (PID: {pid})")

            if open_files:
                print(f"Archivos abiertos por el proceso {pid}:")
                for file in open_files:
                    print(f"Ruta: {file.path}, Descriptor: {file.fd}")
            else:
                print(f"No se encontraron archivos abiertos por el proceso {pid}.")
        except (psutil.NoSuchProcess, psutil.AccessDenied):
            print(f"No se pudo acceder al proceso con PID {random_process.pid}.")
    else:
        print("No se encontraron procesos en ejecuci√≥n.")

# Ejecutar la funci√≥n
list_open_files_random_process()
```
